```python
# 1. Load sentiment files
import pandas as pd

glassdoor_df = pd.read_csv('glassdoor_sentiment.csv')
youtube_df = pd.read_csv('youtube_sentiment.csv')
```

```python
# 2. Standardize column names
glassdoor_df.rename(columns={'summary_cleaned': 'cleaned_text'}, inplace=True)
youtube_df.rename(columns={'transcript_cleaned': 'cleaned_text'}, inplace=True)
```

```python
# 3. Add a platform/source column to each dataset
glassdoor_df['platform'] = 'Glassdoor'
youtube_df['platform'] = 'YouTube'
```

```python
# 4. Combine datasets using pd.concat()
combined_df = pd.concat([glassdoor_df, youtube_df], ignore_index=True)
```

```python
# 5. Save combined dataset
combined_df.to_csv('combined_sentiment_data.csv', index=False)
```
