```python
# Import libraries
import nltk
import string
from nltk.corpus import stopwords
nltk.download('stopwords')

def clean_text(text):
    if pd.isnull(text):
        return ""
    tokens = text.lower().translate(str.maketrans('', '', string.punctuation)).split()
    return ' '.join([word for word in tokens if word not in stopwords.words('english')])
```

```python
# Libraries
import re
from nltk.corpus import stopwords
import nltk
import pandas as pd

nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def clean_text(text):
    if pd.isnull(text):
        return ""
    text = str(text).lower()
    text = re.sub(r'[^a-z\s]', '', text)   # remove punctuation & numbers
    words = text.split()
    words = [word for word in words if word not in stop_words]
    return " ".join(words)
```

```python
# Create cleaned columns
data["summary_cleaned"] = data["summary"].apply(clean_text)
data["advice_to_management_cleaned"] = data["advice_to_management"].apply(clean_text)
data["review_pros_cleaned"] = data["review_pros"].apply(clean_text)
data["review_cons_cleaned"] = data["review_cons"].apply(clean_text)
```

```python
# Compare original vs cleaned text
data[[
    "summary", "summary_cleaned",
    "review_pros", "review_pros_cleaned"
]].head()
```

```python
# Decide what columns to keep
df_final = data[[
    "summary_cleaned",
    "advice_to_management_cleaned",
    "review_pros_cleaned",
    "review_cons_cleaned"
]]
```

```python
# save the final clean dataset
df_final.to_csv("glassdoor_cleaned_text.csv", index=False)
```
