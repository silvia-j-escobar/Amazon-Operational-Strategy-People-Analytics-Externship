```python
import pandas as pd

glassdoor_df = pd.read_csv('glassdoor_sentiment.csv')
youtube_df = pd.read_csv('youtube_sentiment.csv')

youtube_df
#glassdoor_df
```

```python
glassdoor_counts = glassdoor_df['sentiment_label'].value_counts(normalize=True)
youtube_counts = youtube_df['sentiment_label'].value_counts(normalize=True)

print("Glassdoor Sentiment Breakdown:\n", glassdoor_counts)
print("\nYouTube Sentiment Breakdown:\n", youtube_counts)
```

```python
import matplotlib.pyplot as plt

comparison_df = pd.DataFrame({
    'Glassdoor': glassdoor_counts,
    'YouTube': youtube_counts
})

comparison_df.plot(kind='bar')
plt.title("Sentiment Comparison: Glassdoor vs. YouTube")
plt.xlabel("Sentiment")
plt.ylabel("Proportion of Reviews")
plt.xticks(rotation=0)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
